```python
# create a dataframe
dataframe = pd.DataFrame(hospital_df)
dataframe['City']=dataframe['City'].str.lower()
dataframe.drop('SL NO',axis=1,inplace=True)
dataframe
```

```python
def get_recommendation(name):
    
    result_df=dataframe[dataframe['City']==name]
    
    return result_df
    
```
